```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm

# Load the dataset
data_path = '/Users/bishmaybarik/Library/CloudStorage/OneDrive-ShivNadarInstitutionofEminence/Development Economics Term Paper MB/02. Output/Required/final_required.csv'
data = pd.read_csv(data_path)

# Ensure quarter is treated as a categorical variable
data['quarter'] = pd.Categorical(data['quarter'], ordered=True)

# Define the cutoff quarter_centered value (adjusted to 0)
cutoff = 0

# Function to fit linear models
def fit_linear_model(data_subset):
    """
    Fits a linear model and returns predictions and the fitted model.
    """
    X = sm.add_constant(data_subset['quarter_centered'])
    linear_model = sm.OLS(data_subset['share_edu_inc'], X).fit()
    linear_fit = linear_model.predict(X)
    return linear_fit, linear_model

# Plotting function
def plot_linear_graphs(data, religions, cutoff):
    for religion in religions:
        subset = data[data['religion'] == religion]
        
        # Separate data into pre- and post-cutoff
        pre_cutoff = subset[subset['quarter_centered'] <= cutoff]
        post_cutoff = subset[subset['quarter_centered'] >= cutoff]

        # Fit linear model for pre-cutoff
        pre_linear_fit, pre_linear_model = fit_linear_model(pre_cutoff)

        # Predict the value at the cutoff point using the pre-cutoff model
        cutoff_value = pre_linear_model.predict([1, cutoff])[0]

        # Fit linear model for post-cutoff
        X_post = sm.add_constant(post_cutoff['quarter_centered'])
        post_linear_model = sm.OLS(post_cutoff['share_edu_inc'], X_post).fit()

        # Adjust the intercept of the post-cutoff model to ensure continuity at the cutoff
        post_linear_model.params[0] = cutoff_value - post_linear_model.params[1] * cutoff
        post_linear_fit = post_linear_model.predict(X_post)

        # Plot
        plt.figure(figsize=(10, 6))

        # Scatter plot (binned)
        sns.regplot(
            x='quarter_centered', y='share_edu_inc', data=pre_cutoff,
            x_bins=10, scatter_kws={'alpha': 0.5}, label=f'{religion} Pre-Cutoff'
        )
        sns.regplot(
            x='quarter_centered', y='share_edu_inc', data=post_cutoff,
            x_bins=10, scatter_kws={'alpha': 0.5}, label=f'{religion} Post-Cutoff'
        )

        # Add linear fit lines
        plt.plot(pre_cutoff['quarter_centered'], pre_linear_fit, color='blue', label='Linear Fit (Pre)', linestyle='--')
        plt.plot(post_cutoff['quarter_centered'], post_linear_fit, color='red', label='Linear Fit (Post)', linestyle='--')

        # Labels and title
        plt.axvline(x=cutoff, color='black', linestyle='--', label='Cutoff')
        plt.title(f'Share of Education Expenditure by Quarter: {religion}')
        plt.xlabel('Centered Quarter')
        plt.ylabel('Share of Education Expenditure (share_edu_inc)')
        plt.legend()
        plt.grid(True)
        plt.tight_layout()

        # Show plot
        plt.show()

# List of religions in the dataset
religions = data['religion'].unique()

# Plot linear graphs
plot_linear_graphs(data, religions, cutoff)
```
